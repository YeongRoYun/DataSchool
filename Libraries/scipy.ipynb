### Organization
- ```scipy.cluster``` for clustering Algorithms
- ```scipy.constants``` for physical and mathematical constants
- ```scipy.fftpack``` for Fast-Fourier Transformation routines
- ```scipy.integrate``` for integration and ordinary differential equation solvers
- ```scipy.interpolate``` for interpolation and smoothing splines
- ```scipy.io``` for input and output
- ```scipy.linalg``` for Linear Algebra
- ```scipy.ndimage``` for n-Dimension image processing
- ```scipy.odr``` for Orthogonal distance regression
- ```scipy.optimize``` for Optimization and root-finding routines
- ```scipy.signal``` for signal processing
- ```scipy.sparse``` for sparse matrices and associated routines
- ```scipy.spatial``` for spacial data structures and algorithms
- ```scipy.special``` for special functions
- ```scipy.stats``` for statistical distributions and functions
- To view documents, use ```numpy.info(scipy module or functions)```
